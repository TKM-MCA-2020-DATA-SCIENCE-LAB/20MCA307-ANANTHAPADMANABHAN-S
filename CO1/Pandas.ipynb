**Series**

```python
import numpy as np
import pandas as pd
```

```python
labels=['a','b','c','d','e']
list=[100,200,300,400,500]
arr=np.array(list)
d={'a':100,'b':200,'c':300,'d':400,'e':500}
```

```python
pd.Series(data=list)
```

```python
pd.Series(data=list,index=labels)
```

```python
pd.Series(arr,labels)
```

```python
pd.Series(d)
```

```python
s1=pd.Series([1,2,3,4] ,index=["USA","Africa","Italy","India"])
s2=pd.Series([7,9,2,5],index=["Africa","India","China","Bangladesh"])
```

```python
s1+s2
```
